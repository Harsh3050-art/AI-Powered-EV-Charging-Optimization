```python
occupancy_long = occupancy.melt(
    id_vars='timestamp',
    var_name='station_id',
    value_name='occupancy'
)
occupancy_long.head()
```

```python
occupancy_long.info()
```

```python
duration_long = duration.melt(
    id_vars='timestamp',
    var_name='station_id',
    value_name='duration'
)
```

```python
volume_long = volume.melt(
    id_vars='timestamp',
    var_name='station_id',
    value_name='volume'
)
```

```python
price_long = price.melt(
    id_vars='timestamp',
    var_name='station_id',
    value_name='price'
)
```

```python
occupancy_long['station_id'] = occupancy_long['station_id'].astype(int)
duration_long['station_id'] = duration_long['station_id'].astype(int)
volume_long['station_id'] = volume_long['station_id'].astype(int)
price_long['station_id'] = price_long['station_id'].astype(int)
```

```python
time['datetime'] = pd.to_datetime(time[['year','month','day','hour','minute','second']])
time.head()
```

```python
time['timestamp'] = range(1, len(time)+1)
time = time[['timestamp','datetime']]
```

```python
master_df = occupancy_long.merge(
    duration_long,
    on=['timestamp','station_id']
)
```

```python
master_df = master_df.merge(
    volume_long,
    on=['timestamp','station_id']
)
```

```python
master_df = master_df.merge(
    price_long,
    on=['timestamp','station_id']
)
```

```python
master_df = master_df.merge(
    time,
    on='timestamp'
)
```

```python
information = information.rename(columns={'grid':'station_id'})
```

```python
master_df = master_df.merge(
    information,
    on='station_id',
    how='left'
)
```
